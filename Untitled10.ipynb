```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
data = pd.read_excel("cleaned_sales_data (4) (1).xlsx")
print("Dataset Info:")
print(data.info())
print("\nSummary Statistics:")
print(data.describe(include="all"))

# -----------------------------
# 2. Hypothesis Testing
# -----------------------------

# Hypothesis 1: Laptop vs Phone Spending
laptop_spending = data.loc[data["Product"]=="Laptop","Price"]
phone_spending = data.loc[data["Product"]=="Phone","Price"]

if len(laptop_spending) < 2 or len(phone_spending) < 2:
    print("Sample size too small for Laptop vs Phone comparison.")
else:
    t_stat, p_val = stats.ttest_ind(laptop_spending, phone_spending)
    print("\nHypothesis 1: Laptop vs Phone Spending")
    print("T-statistic:", t_stat)
    print("P-value:", p_val)
    if p_val > 0.05:
        print("No significant difference: Laptop and Phone spending are similar.")

# Hypothesis 2: Younger vs Older Customers Spending
younger = data.loc[data["Age"]<30,"Price"]
older = data.loc[data["Age"]>=30,"Price"]

if len(younger) < 2 or len(older) < 2:
    print("Sample size too small for Younger vs Older comparison.")
else:
    t_stat2, p_val2 = stats.ttest_ind(younger, older)
    print("\nHypothesis 2: Younger vs Older Customers Spending")
    print("T-statistic:", t_stat2)
    print("P-value:", p_val2)
    if p_val2 > 0.05:
        print("No significant difference: Age does not impact spending")

# Hypothesis 3: Male vs Female Purchase Counts
male_count = data.loc[data["Gender"]=="Male","Product"].count()
female_count = data.loc[data["Gender"]=="Female","Product"].count()

if male_count < 2 or female_count < 2:
    print("Sample size too small for Male vs Female comparison.")
else:
    chi2, p_val3 = stats.chisquare([male_count,female_count])
    print("\nHypothesis 3: Male vs Female Purchase Counts")
    print("Chi2-statistic:", chi2)
    print("P-value:", p_val3)
    if p_val3 > 0.05:
        print("No significant difference: Gender does not impact purchase frequency")

# -----------------------------
# 3. General Storytelling Visuals
# -----------------------------

# Bar chart: Revenue by Product
plt.figure(figsize=(8,5))
sns.barplot(x="Product", y="Price", data=data, estimator=sum, errorbar=None, hue="Product", legend=False, palette="viridis")
plt.title("Revenue by Product")
plt.xticks(rotation=45)
plt.show()

# Pie chart: Gender Distribution
plt.figure(figsize=(5,5))
data["Gender"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue","lightpink"])
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# Histogram: Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(data["Age"], bins=10, kde=True, color="purple")
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
